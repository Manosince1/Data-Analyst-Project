### Data Preparation
- Dropping irrelevant columns and rows
- Identifying dupplicated rows
- Renaming columns and rows
- Feature Creation

```python
# delete or drop specified columns from the DataFrame
df = df.drop(["Educational Qualifications", "latitude", "longitude", "Pin code", "Output"], axis=1)
df
```

```python
# Identifying duplicated rows
df.drop_duplicates(inplace=True,keep="first")
df
```

```python
df.columns
```

```python
# Renaming columns 
df=df.rename(columns={'Occupation':'Job_Function', 
                   'Marital Status':'Marital_Status',
                   'Monthly Income':'Monthly_Income',
                   'Family size':'Family_size',
                   'Unnamed: 12':'Delivered'              
                  })
df
```

```python
# delete or drop specified columns from the DataFrame
df['Feedback'].unique()
```

```python
df['Feedback'].str.strip(" ")
```

```python
df.head(10)
```

```python
df['Feedback'].unique()
```

```python
df['Feedback']=np.where(df['Feedback']=='Negative ','Negative',df['Feedback'])
df
```

```python
df['Feedback'].unique()
```

```python
df['Monthly_Income'].unique()
```

```python
df['Monthly_Income']=np.where(df['Monthly_Income']=='No Income','$0',df['Monthly_Income'])
df['Monthly_Income']=np.where(df['Monthly_Income']=='Below Rs.10000','$1-10000$',df['Monthly_Income'])
df['Monthly_Income']=np.where(df['Monthly_Income']=='10001 to 25000','$10001-$25000',df['Monthly_Income'])
df['Monthly_Income']=np.where(df['Monthly_Income']=='25001 to 50000','$25001-$50000',df['Monthly_Income'])
df['Monthly_Income']=np.where(df['Monthly_Income']=='More than 50000','$50001-$100000',df['Monthly_Income'])
df
```

### Data Visualisation

```python
df.describe(include="all")
```

```python
# Histogram display the Age by Total Effectif
plt.hist(df["Age"], color="blue", bins = 10)
plt.title("Age", color="Blue", weight="bold")
plt.xlabel("Age", weight="bold")
plt.ylabel("Total Effectif", weight="bold")
plt.grid()
```

```python
# Histograme display the Family Size by Total Effectif
plt.figure(figsize=(10,6))
plt.hist(df["Family_size"], color="Green", bins = 10)
plt.title("Family Size", color="Green", weight="bold")
plt.xlabel("Size", weight="bold")
plt.ylabel("Total Effectif", weight="bold")
plt.grid()
```

```python
data=df
```

```python
# Histograme display the Family Size by Total Effectif
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Monthly_Income", hue="Job_Function", palette="Reds")
plt.title("Salary by Job Occupation", color="Black", weight="bold")
plt.xlabel("Monthly Income", weight="bold")
plt.ylabel("Total Effectif", weight="bold")
plt.grid()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["Feedback"], bins=3)
plt.xlabel("Size", weight="bold")
plt.ylabel("Total Effectif", weight="bold")
plt.title("Feedback Impact", weight="bold")
plt.grid()
```

```python
# Histograme display the Gender by Total Effectif
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Gender", hue="Marital_Status", palette="Blues")
plt.xlabel("Gender", weight="bold")
plt.title("Gender by Status", color="Black", weight="bold")
plt.ylabel("Total Effectif", weight="bold")
```

```python
sns.boxplot(x="Delivered", y="Monthly_Income", data=df)
plt.xlabel("Delivered", weight="bold")
plt.ylabel("Monthly Income", weight="bold")
plt.title("Delivered by Monthly_Income", color="Black", weight="bold")
```
